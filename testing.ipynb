```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_file_name = './Comp-Map-G25-550.jpg'
img = cv2.imread(img_file_name)

plt.figure(dpi=100)
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(dpi=100)
plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
edges = cv2.Canny(gray, 150, 255)

plt.figure(dpi=100)
plt.imshow(edges, cmap='gray', vmin=0, vmax=255)
plt.xticks([])
plt.yticks([])
plt.show()
```
